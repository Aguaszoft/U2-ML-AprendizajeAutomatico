```python
# Ejemplo de Agrupamiento con K-means
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Crear datos de ejemplo para agrupamiento
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Crear modelo K-means con 4 clusters
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)

# Predecir los clusters
y_kmeans = kmeans.predict(X)

# Visualizacin de los clusters
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centros = kmeans.cluster_centers_
plt.scatter(centros[:, 0], centros[:, 1], c='red', s=200, alpha=0.75)
plt.title('Agrupamiento con K-means')
plt.show()
```
